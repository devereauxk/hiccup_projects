```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
N = 10000
N_perjet = 10

pts = np.array([np.random.normal(5, 3, N) for i in range(N_perjet)])
pts_smeared = np.array([np.array([x + np.random.normal(0,2) for x in pt]) for pt in pts])

jet_pt = np.sqrt(np.sum(pts**2, axis=0))
jet_pt_smeared = np.sqrt(np.sum(pts_smeared**2, axis=0))
```

```python
print(pts.shape)
print(pts_smeared.shape)
print(jet_pt.shape)
print(jet_pt_smeared.shape)
```

```python
# particle pt error
binning = np.linspace(0, 15, 100)
all_pts = np.concatenate(pts, axis=0)
all_pts_smeared = np.concatenate(pts_smeared, axis=0)
plt.hist(all_pts, binning, alpha=0.5, label='true')
plt.hist(all_pts_smeared, binning, alpha=0.5, label='smeared')
plt.legend()
plt.xlabel('particle pt')
plt.show()

binning = np.linspace(-2, 2, 100)
plt.hist( (all_pts_smeared - all_pts) / all_pts, binning)
plt.title('particle pt relative error')
plt.show()
```

```python
# jet pt error
binning = np.linspace(10, 30, 100)
plt.hist(jet_pt, binning, alpha=0.5, label='true')
plt.hist(jet_pt_smeared, binning, alpha=0.5, label='smeared')
plt.legend()
plt.xlabel('jet pt')
plt.show()

binning = np.linspace(-2, 2, 100)
plt.hist( (jet_pt_smeared - jet_pt) / jet_pt, binning)
plt.title('jet pt relative error')
plt.show()
```
